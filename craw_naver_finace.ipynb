```python
import pandas as pd
import re
from pprint import pprint

url_tmpl = 'http://companyinfo.stock.naver.com/v1/company/ajax/cF1001.aspx?cmp_cd=%s&fin_typ=%s&freq_typ=%s'
url = url_tmpl % ('005930', '4', 'Y') # 삼성전자, 4(IFRS 연결), Y:년 단위

dfs = pd.read_html(url)
df = dfs[0]
df = df.set_index('주요재무정보')


제거문자열_정규식 = re.compile("[\t\n ]{1,}")
indexes = list(map(lambda x:제거문자열_정규식.split(x.strip())[0], list(df.columns.levels[1])[:-1]))

totalValue = {}
for j in zip(indexes,df.items()):
    datas=j[1][1:][0]
    totalValue[j[0]] = pd.Series(datas.values, index=map(lambda x: x[0], datas.index)) 

totalValueDF = pd.DataFrame(totalValue)

print(totalValue['2016/12']['현금배당성향(%)'])
print(totalValueDF['2016/12']['현금배당성향(%)'])
print(totalValue['2019/12(E)']['자본유보율'])
print(totalValueDF['2019/12(E)']['자본유보율'])

pprint(totalValueDF)
```
